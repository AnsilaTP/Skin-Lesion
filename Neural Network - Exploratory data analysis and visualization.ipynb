The dataset contains 10,015 images of skin lesions, which are classified into 7 categories: 
nv: Melanocytic nevi
mel: Melanoma
bkl: benign keratosis-like lesions
bcc: Basal cell carcinoma
akiec: Actinic keratoses
vasc: Vascular lesions
df: Dermatofibroma

```python
#import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the metadata CSV file into a pandas dataframe
metadata = pd.read_csv('HAM10000_metadata.csv')

```

```python
metadata
```

```python
# Display the first 10 rows of the metadata dataframe
metadata.head(10)
```

```python
# Count the number of images in each diagnosis category
counts = metadata['dx'].value_counts()
```

```python
# Plot the class distribution using a bar chart
plt.figure(figsize=(5,5))
plt.bar(counts.index, counts.values)
plt.title('Class Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the class distribution using a pie chart
plt.figure(figsize=(8,6))
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title('Class Distribution')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(5,5))
plt.hist(metadata['age'], bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Sex Distribution
counts = metadata['sex'].value_counts()
plt.figure(figsize=(5,5))
plt.bar(counts.index, counts.values)
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Lesion Localization
counts = metadata['localization'].value_counts()
plt.figure(figsize=(5,5))
plt.bar(counts.index, counts.values)
plt.title('Lesion Localization')
plt.xlabel('Localization')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
